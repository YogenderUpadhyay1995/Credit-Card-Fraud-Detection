```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv("/content/creditcard.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.describe()
```

#spilting the details

```python
data.Class.value_counts()
```

```python
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
data.groupby('Class').mean()
```

```python
Sample_legit = legit.sample(n=492)
Sample_legit.head()
```

Concatenating or Combining the Data

```python
Credit_card_data = pd.concat([Sample_legit, fraud], axis=0)
```

```python
Credit_card_data.shape
```

```python
Credit_card_data.head()
```

Spliting the data

```python
X= Credit_card_data.drop(columns = "Class")
Y= Credit_card_data["Class"]
```

```python
print(X.shape)
print(Y.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

Logitics Regression Model

```python
Model= LogisticRegression()
```

```python
Model.fit(X_train, Y_train)
```

Evaluation of Model

```python
training_data_prediction = Model.predict(X_train)
```

Accuracy Check

```python
# R Squared Method

r2_error = metrics.r2_score(Y_train, training_data_prediction)
print(r2_error)
```

```python
# Mean absoulte method

Mean_absoulte_method = metrics.mean_absolute_error(Y_train, training_data_prediction)
print(Mean_absoulte_method)
```

Predict on Test data

```python
Training_test_data = Model.predict(X_test)
```

Accuracy check of Test data

```python
# R Squared Method
R2_error_TestData = metrics.r2_score(Y_test, Training_test_data)
print(R2_error_TestData)
```

```python
# mean absoulte Method

Mean_absoulte_method_testdata= metrics.mean_absolute_error(Y_test, Training_test_data)
print(Mean_absoulte_method)
```
